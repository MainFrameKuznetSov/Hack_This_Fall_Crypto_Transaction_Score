```python
import pandas as pd
from collections import defaultdict as dd

path='Refined_Dataset.csv'
frm=pd.read_csv(path,usecols=['From'])
to=pd.read_csv(path,usecols=['To'])
err=pd.read_csv(path,usecols=['Err'])

u=frm['From'].tolist()
v=to['To'].tolist()
wt=err['Err'].tolist()

adj_list=dd(list)

for i in range(len(u)):
    adj_list[u[i]].append((v[i],wt[i]))

data=[]
for key,value in adj_list.items():
    for it in value:
        data.append({'From':key,'To':it[0],'Err':it[1]})

temp=pd.DataFrame(data)
temp.to_csv('Adjacency_List.csv',index=False)

print(len(data))
print(data[0])
```

```python
from google.colab import files
import pandas as pd
from collections import defaultdict as dd
den=dd(int)
num=dd(int)
scam=dd(int)
legit=dd(int)
for node,edge in adj_list.items():
  #print(edge)
  for _,wt in edge:
    num[node]+=int(wt[1])
    den[node]+=1
    if(int(wt[1])==1):
      scam[node]+=1
    else:
      legit[node]+=1


data=[]
c=0
for node in adj_list:
  prob=num[node]/den[node]
  if(prob>0):
    c+=1
    str=node[2:len(node)-2]
    data.append({"Wallet":str,"No. of transaction":den[node],"Scams":scam[node],"Legit":legit[node],"Score":prob})

map={entry["Wallet"]:entry["Scams"] for entry in data}
for i,j in map.items():
    print(f"{i}->{j}")

temp=pd.DataFrame(data)
temp.to_csv("Scammers.csv",index=False)
print(temp)
print(c)
#files.download("Scammers.csv")
```
